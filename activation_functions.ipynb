comparson between various activation functions

```
import numpy as np
from numpy import array
import pandas as pd
import matplotlib.pyplot as plt
from numpy import random
from google.colab import files
```

Step or hardlim

```
def step(x):
  return(np.where(x>=0,1,-1))

def impulse(x):
  return(np.where(x==0,1,0))

x = np.arange(-10,10,0.1)
y = step(x)

plt.plot(x,y,'r-.',linewidth=3.0)
plt.plot(x,impulse(x),'g-',linewidth=3.0)
plt.title('Step')
plt.xlabel('x')
plt.xlabel('y')
plt.grid()
plt.show()
```

Sigmoid activation function

```
def sig(x):
  return(np.exp(x)/(1+np.exp(x)))

def sig_derivative(x):
  return(sig(x)*(1-sig(x)))

plt.plot(x,sig(x),'r-', linewidth=3.0)
plt.plot(x,sig_derivative(x),'g-', linewidth=3.0)
plt.xlabel('x')
plt.ylabel('sig(x)')
plt.grid()
plt.show()
```

```
#softmax: normalized exponential function
# it  Calculate the softmax for the give inputs (array)
def softmax(x):
  return(np.exp(x)/np.exp(x).sum())

x = [2,3,5,6]
print(softmax(x))
print(softmax(x).sum())
```

```
# tanh(x)
x = np.arange(-10,10,0.1)

def tanh(x):
  return((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))

def tanh_derivative(x):
  return(1-tanh(x)**2)

plt.plot(x,tanh(x),'r-', linewidth=3.0)
plt.plot(x,tanh_derivative(x),'g-', linewidth=3.0)
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid()
plt.show()
```

```
# ReLU: Rectifier 
x = np.arange(-10,10,0.1)

def relu_(x):
  return(np.maximum(x,0))

def softplus_(x):
  return(np.log((1+np.exp(x))))

def relu_derivative(x):
  return(np.where(x>=0,1,0))

def softplus_derivative(x):
  return(1/(1+np.exp(-x)))

plt.plot(x, relu_(x),'r-', linewidth=3.0)
plt.plot(x, softplus_(x), 'b-', linewidth=3.0)
plt.plot(x, relu_derivative(x),'r:', linewidth = 3.0)
plt.plot(x, softplus_derivative(x),'b:',linewidth = 3.0 )
plt.grid()
plt.show()
```
